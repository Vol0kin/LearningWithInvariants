```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('glass_multiple_fits.csv')
df
```

```python
df['min_accuracy'] = df[[f'accuracy_{i}' for i in range(1, 6)]].min(axis=1)
df['max_accuracy'] = df[[f'accuracy_{i}' for i in range(1, 6)]].max(axis=1)
df['avg_accuracy'] = df[[f'accuracy_{i}' for i in range(1, 6)]].mean(axis=1)
df
```

```python
def generate_bar_chart(df, y_variable, y_title):
    return  alt.Chart(df).mark_bar(width=8).encode(
        x=alt.X(
            'model:N',
            axis=alt.Axis(
                title=None,
                labels=False
            )
        ),
        y=alt.Y(
            y_variable,
            title=y_title
        ),
        color='model:N',
        tooltip=[alt.Tooltip('avg_accuracy:Q', format=".3f"), 'model:N'],
        column=alt.Column(
            'size:N',
            spacing=30,
            title='Percentage of training data',
        ),
    ).properties(
        width=100,
        height=200,
    )
```

```python
generate_bar_chart(df, 'avg_accuracy:Q', 'Average accuracy')
```

```python
generate_bar_chart(df, 'min_accuracy:Q', 'Minimum accuracy')
```

```python
generate_bar_chart(df, 'max_accuracy:Q', 'Maximum accuracy')
```
