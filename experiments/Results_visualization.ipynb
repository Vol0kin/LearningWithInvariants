```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('results.csv')
df
```

```python
bars = alt.Chart(df).mark_bar(width=8).encode(
    x=alt.X(
        'model:N',
        axis=alt.Axis(
            title=None,
            labels=False
        )
    ),
    y=alt.Y(
        'accuracy:Q',
        title='Accuracy'
    ),
    color='model:N',
    tooltip=[alt.Tooltip('accuracy:Q', format=".3f"), 'model:N'],
    column=alt.Column(
        'size:N',
        spacing=15,
        title='Percentage of training data',
    ),
    row=alt.Row(
        'problem:N',
        title='Problem',
        header=alt.Header(
            labelAngle=0,
            labelAlign='left'
        ),
        spacing=40),
).properties(
    width=75,
    height=150,
)

bars
```
